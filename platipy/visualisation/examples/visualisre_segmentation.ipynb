```python
import numpy as np
import SimpleITK as sitk
import matplotlib.pyplot as plt

%matplotlib notebook
%matplotlib notebook

from platipy.visualisation import vis_tools
```

```python
image = sitk.ReadImage('../../segmentation/tests/data/dynamic/lung/LCTSC-Train-S1-008/CT.nii.gz')

structure_name_list = ["Heart","Esophagus","Lung_L","Lung_R", "SpinalCord"]
contours = {s:sitk.ReadImage(f'../../segmentation/examples/Auto_{s}.nii.gz') for s in structure_name_list}
```

```python
fig, ax, cut = vis_tools.displaySlice(image);
```

```python
for s in structure_name_list:
    fig = vis_tools.overlayContour(contours[s], fig, ax, cut, plt.cm.Spectral)
```

```python
plt.draw(); fig
```

```python
fig.canvas.draw()
```
